```python
import json
from bs4 import BeautifulSoup
from urllib.parse import urljoin

BASE_URL = "https://www.thecrag.com"  # 需要绝对链接就用它拼接


def norm_text(s: str) -> str:
    return " ".join((s or "").split()).strip()


def parse_li(li, base_url: str = BASE_URL):
    """
    解析一个 <li> 节点，返回 dict:
    {
      "name": "...",
      "href": "https://www.thecrag.com/...",
      "nid": "...",
      "subtype": "...",
      "children": [...]
    }
    """
    # 取当前 li 的“直接子 a”（避免误拿到更深层子菜单里的 a）
    a = li.find("a", recursive=False)
    if a is None:
        # 有些 li 可能只起结构作用，尝试向下找第一个 a
        a = li.find("a")

    if a is None:
        return None

    href = a.get("href")
    href_abs = urljoin(base_url, href) if href else None

    nid = a.get("data-nid")
    subtype = a.get("data-subtype")

    # 优先使用 primary-node-name 的文本，否则用 a 的文本
    name_span = a.find("span", class_="primary-node-name")
    name = norm_text(name_span.get_text()) if name_span else norm_text(a.get_text())

    node = {"name": name, "href": href_abs, "nid": nid, "subtype": subtype, "children": []}

    # 取当前 li 的“直接子 ul”（子菜单）
    child_ul = li.find("ul", recursive=False)
    if child_ul is not None:
        for child_li in child_ul.find_all("li", recursive=False):
            child_node = parse_li(child_li, base_url=base_url)
            if child_node is not None:
                node["children"].append(child_node)

    return node


def extract_regions_tree(html: str, base_url: str = BASE_URL):
    soup = BeautifulSoup(html, "html.parser")

    # 锚点：ul.embed-menu
    root_ul = soup.select_one("ul.embed-menu")
    if root_ul is None:
        raise ValueError("未找到 ul.embed-menu；请确认传入的是包含该节点的 HTML。")

    result = []
    for li in root_ul.find_all("li", recursive=False):
        node = parse_li(li, base_url=base_url)
        if node is not None:
            result.append(node)

    return result


html = html_element
tree = extract_regions_tree(html)
print(json.dumps(tree, ensure_ascii=False, indent=2))
```

```python
def remove_citation_fields(obj, *, case_insensitive=True):
    """
    Recursively remove keys whose name contains 'citation'.

    Args:
        obj: dict | list | primitive
        case_insensitive: whether key matching is case-insensitive

    Returns:
        Cleaned object (same type as input)
    """
    if isinstance(obj, dict):
        new_obj = {}
        for k, v in obj.items():
            key_cmp = k.lower() if case_insensitive else k
            if "citation" in key_cmp or "href" in key_cmp or "nid" in key_cmp:
                continue
            if "subtype" in obj and obj["subtype"] == "gym":
                continue
            if "gym" in obj.get("name", "").lower():
                continue
            new_obj[k] = remove_citation_fields(v, case_insensitive=case_insensitive)
        return new_obj

    if isinstance(obj, list):
        return [remove_citation_fields(x, case_insensitive=case_insensitive) for x in obj]

    # primitive types: str, int, float, bool, None
    return obj


output_json = remove_citation_fields(json.load(open("../resources/routes.json", "r", encoding="utf-8")))

with open("../resources/regions_tree.json", "w", encoding="utf-8") as f:
    json.dump(output_json, f, ensure_ascii=False, indent=2)
```
